```python
import fitz  # PyMuPDF for PDF text extraction
import spacy
from spacy.matcher import PhraseMatcher
from skillNer.skill_extractor_class import SkillExtractor
from skillNer.general_params import SKILL_DB

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to read a text job description
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        return file.read()
```

```python
from google.colab import files

# Upload the resume (PDF) and job description (TXT)
uploaded = files.upload()

# Separate uploaded files
pdf_path = next(key for key in uploaded if key.endswith('.pdf'))
txt_path = next(key for key in uploaded if key.endswith('.txt'))
```

```python
# Load spaCy model
nlp = spacy.load("en_core_web_lg")

# Initialize SkillExtractor
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)

# Extract text from the resume
resume_text = extract_text_from_pdf(pdf_path)

# Extract text from the job description
job_description_text = extract_text_from_txt(txt_path)

# Annotate and extract skills from both texts
resume_annotations = skill_extractor.annotate(resume_text)
job_description_annotations = skill_extractor.annotate(job_description_text)

# Get the extracted skills
resume_skills = set(skill["doc_node_value"].lower() for skill in resume_annotations["results"]["full_matches"])
job_description_skills = set(skill["doc_node_value"].lower() for skill in job_description_annotations["results"]["full_matches"])

# Remove "com" and duplicates
resume_skills = {skill for skill in resume_skills if skill != "com"}
job_description_skills = {skill for skill in job_description_skills if skill != "com"}

# Find matched skills
matched_skills = resume_skills.intersection(job_description_skills)
```

```python
print("Extracted Skills from Resume:")
print(", ".join(resume_skills))

print("\nExtracted Skills from Job Description:")
print(", ".join(job_description_skills))

print("\nMatched Skills:")
print(", ".join(matched_skills))
```

```python
import spacy
import fitz  # For extracting text from PDFs

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to read text from a TXT file
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Function to extract skills from text using heuristic rules
def extract_skills(text):
    doc = nlp(text)
    # Example heuristic: Match potential skill-related terms using simple criteria
    skills = [ent.text for ent in doc.ents if ent.label_ in {'ORG', 'PRODUCT', 'WORK_OF_ART', 'LANGUAGE'}]
    return set(skills)  # Convert to a set to remove duplicates
```

# Combination of SkillNER and Spacey

```python
import fitz  # For extracting text from PDFs
import spacy
from spacy.matcher import PhraseMatcher
from skillNer.skill_extractor_class import SkillExtractor
from skillNer.general_params import SKILL_DB

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to read text from a TXT file
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Function to extract skills using SkillNER
def extract_skills_skillner(text, skill_extractor):
    annotations = skill_extractor.annotate(text)
    skills = {skill["doc_node_value"].lower() for skill in annotations["results"]["full_matches"]}
    return skills

# Function to extract skills using spaCy
def extract_skills_spacy(text):
    doc = nlp(text)
    # Example heuristic: Match potential skill-related terms using entity labels
    skills = {ent.text.lower() for ent in doc.ents if ent.label_ in {'ORG', 'PRODUCT', 'WORK_OF_ART', 'LANGUAGE'}}
    return skills
```

```python
from google.colab import files

uploaded = files.upload()

# Identify uploaded files
pdf_path = next(key for key in uploaded if key.endswith('.pdf'))
txt_path = next(key for key in uploaded if key.endswith('.txt'))
```

```python
# Initialize SkillNER
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)

# Extract text from the uploaded files
resume_text = extract_text_from_pdf(pdf_path)
jd_text = extract_text_from_txt(txt_path)

# Extract skills using SkillNER
resume_skills_skillner = extract_skills_skillner(resume_text, skill_extractor)
jd_skills_skillner = extract_skills_skillner(jd_text, skill_extractor)

# Extract skills using spaCy
resume_skills_spacy = extract_skills_spacy(resume_text)
jd_skills_spacy = extract_skills_spacy(jd_text)

# Combine skills from both methods
resume_skills = resume_skills_skillner.union(resume_skills_spacy)
jd_skills = jd_skills_skillner.union(jd_skills_spacy)

# Remove noise: Exclude specific terms and duplicates
noise_terms = {"com"}  # Add any additional terms to exclude
resume_skills = {skill for skill in resume_skills if skill not in noise_terms}
jd_skills = {skill for skill in jd_skills if skill not in noise_terms}

# Find matching skills
matched_skills = resume_skills.intersection(jd_skills)
```

```python
print("Extracted Skills from Resume:")
print(", ".join(sorted(resume_skills)))

print("\nExtracted Skills from Job Description:")
print(", ".join(sorted(jd_skills)))

print("\nMatched Skills:")
print(", ".join(sorted(matched_skills)))
```

```python
# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Function to read text from a TXT file
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Function to extract skills using SkillNER
def extract_skills_skillner(text, skill_extractor):
    annotations = skill_extractor.annotate(text)
    skills = {skill["doc_node_value"].lower() for skill in annotations["results"]["full_matches"]}
    return skills

# Function to extract skills using spaCy
def extract_skills_spacy(text):
    doc = nlp(text)
    skills = {ent.text.lower() for ent in doc.ents if ent.label_ in {'ORG', 'PRODUCT', 'WORK_OF_ART', 'LANGUAGE'}}
    return skills

# Function to preprocess skills (remove duplicates and noise)
def preprocess_skills(skills):
    noise_terms = {"com"}  # Add more noise terms as needed
    return {skill for skill in skills if skill not in noise_terms}

# Function to calculate SBERT similarity between two lists of skills
def calculate_sbert_similarity(skills_1, skills_2):
    if not skills_1 or not skills_2:
        return 0.0  # Return 0 if either list is empty
    # Generate embeddings
    embeddings_1 = sbert_model.encode(list(skills_1), convert_to_tensor=True)
    embeddings_2 = sbert_model.encode(list(skills_2), convert_to_tensor=True)
    # Compute similarity
    similarity_score = util.cos_sim(embeddings_1, embeddings_2).mean().item()
    return similarity_score
```

```python
import os
import fitz  # PyMuPDF for PDF text extraction
import spacy
from spacy.matcher import PhraseMatcher
from skillNer.skill_extractor_class import SkillExtractor
from skillNer.general_params import SKILL_DB
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Load spaCy and SBERT models
nlp = spacy.load("en_core_web_sm")
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    """
    Extracts text from a PDF. Skips if the PDF contains only images or is empty.
    """
    text = ""
    try:
        with fitz.open(pdf_path) as doc:
            for page in doc:
                page_text = page.get_text()
                if page_text.strip():  # Check if the page has text
                    text += page_text
    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
        return None

    if not text.strip():  # Check if the entire PDF is empty
        print(f"Warning: No extractable text found in {pdf_path}. Skipping...")
        return None
    return text


# Function to read text from a TXT file
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Function to extract skills using SkillNER
def extract_skills_skillner(text, skill_extractor):
    annotations = skill_extractor.annotate(text)
    skills = {skill["doc_node_value"].lower() for skill in annotations["results"]["full_matches"]}
    return skills

# Function to extract skills using spaCy
def extract_skills_spacy(text):
    doc = nlp(text)
    skills = {ent.text.lower() for ent in doc.ents if ent.label_ in {'ORG', 'PRODUCT', 'WORK_OF_ART', 'LANGUAGE'}}
    return skills

# Function to preprocess skills (remove duplicates and noise)
def preprocess_skills(skills):
    noise_terms = {"com"}  # Add more noise terms as needed
    return {skill for skill in skills if skill not in noise_terms}

# Function to calculate SBERT similarity between two lists of skills
def calculate_sbert_similarity(skills_1, skills_2):
    if not skills_1 or not skills_2:
        return 0.0  # Return 0 if either list is empty
    # Generate embeddings
    embeddings_1 = sbert_model.encode(list(skills_1), convert_to_tensor=True)
    embeddings_2 = sbert_model.encode(list(skills_2), convert_to_tensor=True)
    # Compute similarity
    similarity_score = util.cos_sim(embeddings_1, embeddings_2).mean().item()
    return similarity_score
```

```python
import re

def preprocess_text_for_skillner(text):
    """
    Preprocess text to clean up special characters and non-standard content
    that could cause SkillNER to fail.
    """
    # Remove special characters, excessive whitespace, and non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)  # Remove non-ASCII characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove excessive whitespace
    return text
```

```python
# Function to extract skills using SkillNER
def extract_skills_skillner(text, skill_extractor):
    """
    Extract skills using SkillNER, with added error handling.
    """
    try:
        cleaned_text = preprocess_text_for_skillner(text)  # Preprocess text
        annotations = skill_extractor.annotate(cleaned_text)
        skills = {skill["doc_node_value"].lower() for skill in annotations["results"]["full_matches"]}
        return skills
    except Exception as e:
        print(f"Error during SkillNER annotation: {e}")
        return set()  # Return an empty set if an error occurs
```

```python
# Path to the folder containing resumes and job description
resumes_folder = "/content/resumes"  # Folder containing PDF resumes
job_description_file = "/content/job_description.txt"  # TXT file containing job description

# Initialize SkillNER
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)

# Read and extract text from job description
job_description_text = extract_text_from_txt(job_description_file)

# Extract keywords from job description
jd_skills_skillner = extract_skills_skillner(job_description_text, skill_extractor)
jd_skills_spacy = extract_skills_spacy(job_description_text)
jd_skills = preprocess_skills(jd_skills_skillner.union(jd_skills_spacy))

# Prepare a DataFrame to store results
results = []

# Process each resume in the folder
# Process each resume in the folder
# Process each resume in the folder
for resume_file in os.listdir(resumes_folder):
    if resume_file.endswith(".pdf"):  # Ensure it's a PDF file
        resume_path = os.path.join(resumes_folder, resume_file)
        print(f"Processing {resume_file}...")

        # Extract text from PDF
        resume_text = extract_text_from_pdf(resume_path)

        # Skip if no valid text is extracted
        if not resume_text:
            print(f"Skipping {resume_file} due to lack of text.")
            continue

        # Extract keywords from resume
        try:
            resume_skills_skillner = extract_skills_skillner(resume_text, skill_extractor)
            resume_skills_spacy = extract_skills_spacy(resume_text)
            resume_skills = preprocess_skills(resume_skills_skillner.union(resume_skills_spacy))
        except Exception as e:
            print(f"Error processing {resume_file}: {e}")
            continue

        # Match skills and calculate SBERT similarity
        matched_skills = resume_skills.intersection(jd_skills)
        sbert_similarity = calculate_sbert_similarity(resume_skills, jd_skills)

        # Append results to the DataFrame
        results.append({
            "Resume": resume_file,
            "Resume Keywords": ", ".join(sorted(resume_skills)),
            "Job Description Keywords": ", ".join(sorted(jd_skills)),
            "Matched Keywords": ", ".join(sorted(matched_skills)),
            "Similarity Score": round(sbert_similarity, 2)
        })
```

```python
# Convert results to a Pandas DataFrame
results_df = pd.DataFrame(results)

# Save the DataFrame to an Excel file
output_file = "resume_matching_results.xlsx"
results_df.to_excel(output_file, index=False, engine="openpyxl")

print(f"Results saved to {output_file}")9
```
